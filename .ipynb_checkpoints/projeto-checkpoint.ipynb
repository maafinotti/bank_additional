```python
#gráfico para visualização da satisfação por categorias
fig = plt.subplots(figsize = (9,6))
contagem_categorias = df['y'].value_counts()
plt.bar(contagem_categorias.index, contagem_categorias.values)
for i, count in enumerate(contagem_categorias.values):
    plt.text(i, count, str(count), ha='center', va='bottom')
plt.title('O cliente vai sobrescever um depósito a prazo?', fontsize = 18, style = "italic", loc = "center", pad = 15)
plt.ylabel('Contagem', fontsize = 12, style = "italic", labelpad = 12)
plt.grid()
```

```python
df_sem_strings = df.select_dtypes(exclude=['object'])
```

```python
#visualização da distribuição dos dados numéricos
df_sem_strings.hist(figsize=(10, 8), color = "#902B5C")
plt.tight_layout()
```

```python
#gráfico de calor de correlação
correlacao = df_sem_strings.corr()
plt.figure(figsize=(15, 8))
sns.heatmap(correlacao, annot=True, cmap="coolwarm")
plt.title('Matriz de Correlação', fontsize = 18, style = "italic", loc = "center", pad = 15)
```
